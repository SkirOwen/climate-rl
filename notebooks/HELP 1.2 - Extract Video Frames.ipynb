```python
import cv2
import matplotlib.pyplot as plt
```

```python
BASE_DIR = "/gws/nopw/j04/ai4er/users/pn341/climate-rl"
VIDEO_PATH = f"{BASE_DIR}/videos/rce-v0-optim-L-10k_2024-05-31_14-30/RadiativeConvectiveModel-v0__dpg_torch__1__1717162290/rl-video-episode-19.mp4"
```

```python
def FrameCapture(path):
    vidObj = cv2.VideoCapture(path)
    count = 0
    success = 1
    while success:
        success, image = vidObj.read()
        if success:
            RGB_img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        count += 1
    return RGB_img
```

```python
img = FrameCapture(VIDEO_PATH)

fig = plt.figure(figsize=(27, 9))
plt.imshow(img)
plt.axis("off")

plt.savefig(
    f"{BASE_DIR}/tmp/dpg_rce-v0-optim-L-10k.png",
    format="png",
    bbox_inches="tight",
    dpi=300,
)
```
