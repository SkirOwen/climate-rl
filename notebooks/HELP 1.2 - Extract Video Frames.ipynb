```python
def FrameCapture(path):
    vidObj = cv2.VideoCapture(path)
    count = 0
    success = 1
    while success:
        success, image = vidObj.read()
        if success:
            RGB_img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        count += 1
    return RGB_img
```

```python
img = FrameCapture(VIDEO_PATH)

plt.imshow(img)
plt.axis("off");
```
